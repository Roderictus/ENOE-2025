```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set_theme(style="whitegrid")

# Leer tu base de datos final
direccion = 'C:/2025_Proyectos/ENOE 2025/ENOE_Series_Tiempo_20260330.csv'
df = pd.read_csv(direccion)

# Crear columna de periodo (e.g., "2005-T1") para el eje X
df['Periodo'] = df['year'].astype(int).astype(str) + '-T' + df['quarter'].astype(int).astype(str)

# Asegurar orden cronológico
df = df.sort_values(by=['year', 'quarter']).reset_index(drop=True)

# Inicializar figura
plt.figure(figsize=(14, 8))

# Graficar niveles educativos (ingresos > 0, deflactados)
# Usamos una paleta de colores de azul claro a oscuro según la escolaridad
plt.plot(df['Periodo'], df['def_ing_mensual_ing_Primaria'], label='Primaria', linewidth=2, color='#5ba0d0')
plt.plot(df['Periodo'], df['def_ing_mensual_ing_Secundaria'], label='Secundaria', linewidth=2, color='#2878b5')
plt.plot(df['Periodo'], df['def_ing_mensual_ing_Preparatoria'], label='Preparatoria', linewidth=2, color='#0f4c81')
plt.plot(df['Periodo'], df['def_ing_mensual_ing_Profesional'], label='Profesional (Universitaria)', linewidth=2.5, color='#17202a')

# Graficar Salario Mínimo Mensual Real (Destacado en rojo)
plt.plot(df['Periodo'], df['def_sm_mensual_estimado'], label='Salario Mínimo Mensual (Real)', 
         linewidth=3, color='#c0392b', linestyle='--')

# Títulos y etiquetas
plt.title('Evolución del Ingreso Mensual Real por Escolaridad vs Salario Mínimo (2005-2025)\n(Pesos constantes base 2018, población ocupada con ingresos > 0)', fontsize=15, pad=15)
plt.xlabel('Año y Trimestre', fontsize=12)
plt.ylabel('Ingreso Mensual Real (MXN Base 2018)', fontsize=12)

# Filtrar etiquetas del eje X para no saturar (Mostramos solo el T1 de cada año)
ticks_to_show = df['Periodo'][df['quarter'] == 1]
plt.xticks(ticks_to_show, rotation=45) 

# Ajustes de la leyenda
plt.legend(title='Nivel Educativo / Referencia', bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=11, title_fontsize=12)
plt.tight_layout()

# Guardar la gráfica en alta calidad
plt.savefig('evolucion_salarial_vs_sm.png', dpi=300, bbox_inches='tight')
plt.show()
```
